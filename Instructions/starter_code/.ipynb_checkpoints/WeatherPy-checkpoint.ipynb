```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import json 
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
# creating a random point 
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key 
# List of city data
city_data = []
city = "Atlanta"

# Create endpoint URL with each city
city_url = url + "&q=" + city


city_weather = requests.get(city_url).json()
print(json.dumps(city_weather, indent = 4, sort_keys = True))
```

```python

city_name = city_weather["name"]
latitude = city_weather["coord"]["lat"]
longitude = city_weather["coord"]["lon"]
temperature = city_weather["main"]["temp"]
humidity = city_weather["main"]["humidity"]
cloudiness = city_weather["clouds"]["all"]
windspeed = city_weather["wind"]["speed"]
country = city_weather["sys"]["country"]
date = city_weather["dt"]

print(f"City_name = {city_name}")
print(f"Latitude = {latitude}")
print(f"Longitude = {longitude}")
print(f"Temperature = {temperature}")
print(f"Humidity = {humidity}")
print(f"Cloudiness = {cloudiness}")
print(f"Windspeed = {windspeed}")
print(f"Country = {country}")
print(f"Date = {date}")
```

```python


# Append the City information into city_data list
city_data.append({"city": city_name, "lat": latitude, "lng": longitude, "temp": temperature, "humidity": humidity, "cloudiness": cloudiness, "windspeed": windspeed, "country": country})
city_data
```

```python
my_weatherdata_df = pd.DataFrame(city_data)
my_weatherdata_df
```

```python
# For Loop 

url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key 
city_data = []

for city in cities:
    city_url = url + "&q=" + city
    
    try:
        print(city)
        city_weather = requests.get(city_url).json()
        city_name = city_weather["name"]
        latitude = city_weather["coord"]["lat"]
        longitude = city_weather["coord"]["lon"]
        temperature = city_weather["main"]["temp"]
        humidity = city_weather["main"]["humidity"]
        cloudiness = city_weather["clouds"]["all"]
        windspeed = city_weather["wind"]["speed"]
        country = city_weather["sys"]["country"]
        date = city_weather["dt"]
        
        city_data.append({"city": city_name, "lat": latitude, "lng": longitude, "temp": temperature, "humidity": humidity, "cloudiness": cloudiness, "windspeed": windspeed, "country": country, "date": date})
    except: 
        print("API call failed for current city")
    
my_weatherdata_df = pd.DataFrame(city_data)
print(my_weatherdata_df)
```

## Linear Regression

```python


# OPTIONAL: Create a function to create Linear Regression plots

# Defining function to be used multiple times in the future
def plot_linear_regression(x_values, y_values, title, x_label, y_label, text_coordinates):
    # Run regression on hemisphere weather data.
    (slope, intercept, r_value, p_value, std_err) = linregress(x_values, y_values)
    # Calculate the regression line "y values" from the slope and intercept.
    regress_values = x_values * slope + intercept
    # Get the equation of the line.
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    # Create a scatter plot and plot the regression line.
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r")
    # Annotate the text for the line equation.
    plt.annotate(line_eq, text_coordinates, fontsize=15, color="red")
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.show()

```

```python

# Create Northern (greater than 0) and Southern Hemisphere (less than 0)DataFrames


nothern_hem = my_weatherdata_df[my_weatherdata_df["lat"] > 0]
southern_hem = my_weatherdata_df[my_weatherdata_df["lat"] < 0]


```


####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression



```python
x = nothern_hem["lat"]
y = nothern_hem["temp"]

plot_linear_regression(x, y, 
                       "Nothern Hemisphere - Temp vs. Lat",
                       "Latitude", 
                       "Temperature", 
                       (0,20))
```
